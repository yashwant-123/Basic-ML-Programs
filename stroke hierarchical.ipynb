```python
import sklearn
import seaborn as sns
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import adjusted_rand_score
from scipy.cluster.hierarchy import linkage
from sklearn import preprocessing
from scipy.cluster.hierarchy import dendrogram
import matplotlib.pyplot as plt
```

```python
stroke=pd.read_csv("healthcare-dataset-stroke-data.csv")
stroke=stroke.fillna(stroke.mean())
stroke=pd.get_dummies(stroke, columns=['gender', 'ever_married','work_type','Residence_type','smoking_status'])
y_r=stroke.stroke
x_r=stroke.drop(['stroke','id'],axis='columns')
print(x_r)
print(y_r)
```

```python


ward = AgglomerativeClustering(n_clusters=3)
ward_pred = ward.fit_predict(x_r)
complete = AgglomerativeClustering(n_clusters=3, linkage='complete')
complete_pred = complete.fit_predict(x_r)
avg = AgglomerativeClustering(n_clusters=3, linkage='average')


ward_ar_score = adjusted_rand_score(y_r, ward_pred)
complete_ar_score = adjusted_rand_score(y_r, complete_pred)
avg_ar_score = adjusted_rand_score(y_r, ward_pred)
print( "Scores: \nWard:", ward_ar_score,"\nComplete: ", complete_ar_score, "\nAverage: ", avg_ar_score)


normalized_X = preprocessing.normalize(x_r)
normalized_X[:10]
linkage_type = 'ward'

linkage_matrix = linkage(normalized_X, linkage_type)

plt.figure(figsize=(22,18))

dendrogram(linkage_matrix)

plt.show()
```
